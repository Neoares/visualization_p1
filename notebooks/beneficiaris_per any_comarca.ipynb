```python
beneficiaris = pd.read_csv(os.path.join(ROOT_PATH, 'beneficiaris.csv'))
print(beneficiaris.shape)
beneficiaris.head()
```

```python
import matplotlib.pyplot as plt
import altair as alt
import numpy as np
```

```python


highlight = alt.selection(type='single', on='mouseover',
                          fields=['NomComarca'], nearest=True)

nearest = alt.selection(type='single', nearest=True, on='mouseover',
                        fields=['x'], empty='none')

base = alt.Chart(beneficiaris).encode(
    x='Any:T',
    y='Beneficiaris',
    #y=alt.Y('Beneficiaris:Q', scale=alt.Scale(type='log', base=10)),
    color=alt.Color('NomComarca', legend=alt.Legend(title="Comarca")),
    #tooltip = ['NomComarca', 'Beneficiaris']
)

points = base.mark_circle().encode(
    opacity=alt.value(0)
).add_selection(
    highlight
).properties(
    width=600
)


lines = base.mark_line().encode(
    size=alt.condition(~highlight, alt.value(1), alt.value(4))
).interactive()

# Draw text labels near the points, and highlight based on selection
text = lines.mark_text(align='left', dx=5, dy=-5).encode(
    text=alt.condition(nearest, 'Beneficiaris:Q', alt.value(' '))
)

alt.layer(points, lines).configure_axis(
   grid=False
).configure_view(
   strokeOpacity=0.5
)
```
