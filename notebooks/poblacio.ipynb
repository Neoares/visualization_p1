```python
import pandas as pd
import numpy as np
import altair as alt
from altair import datum
#import datetime
#import pickle

import sys
sys.path.append("../library/")
from load_data import *
%load_ext autoreload
%autoreload 2

PATH_DATA = '../data/'
# to enable Altair in the classic notebook
alt.renderers.enable('notebook')
```

```python
df_poblacio_1 = pd.read_html(
    PATH_DATA + "poblacio_2004_2008.xls", 
    header=0,
    encoding="windows-1252",
    na_values='..',
    thousands="."
)[0]
df_poblacio_1.rename(columns={"Unnamed: 0":"NomComarca"}, inplace=True)
df_poblacio_1.drop("Unnamed: 6", axis=1, inplace=True)
df_poblacio_1
```

```python
df_poblacio_2 = pd.read_html(
    PATH_DATA + "poblacio_2009_2013.xls", 
    header=0,
    encoding="utf-8",
    na_values='..',
    thousands="."
)[0]
df_poblacio_2.rename(columns={"Unnamed: 0":"NomComarca_2"}, inplace=True)
df_poblacio_2.drop("Unnamed: 6", axis=1, inplace=True)
df_poblacio_2
```

```python
df_poblacio_3 = pd.read_html(
    PATH_DATA + "poblacio_2014_2018.xls", 
    header=0,
    encoding="windows-1252",
    na_values='..',
    thousands="."
)[0]
df_poblacio_3.rename(columns={"Unnamed: 0":"NomComarca_3"}, inplace=True)
df_poblacio_3.drop("Unnamed: 6", axis=1, inplace=True)
df_poblacio_3
```

```python
df_poblacio = pd.concat([df_poblacio_1, df_poblacio_2, df_poblacio_3], axis=1, sort=False)
df_poblacio.drop("NomComarca_2", axis=1, inplace=True)
df_poblacio.drop("NomComarca_3", axis=1, inplace=True)
df_poblacio.to_csv(PATH_DATA + "poblacio.csv",index=False)
df_poblacio
```
