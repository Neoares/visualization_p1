```
import altair as alt
import pandas as pd
```

```
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```
food_file = 'drive/My Drive/Dashboard visualization/data/foodpandas.csv'
```

```
food_df = pd.read_csv(food_file,sep=',')
```

```
food_df.head()
```

```
alt.data_transformers.enable('default', max_rows=None)
```

```
alt.Chart(food_df
).mark_line(
    point=True
).encode(
    alt.X('year(Any):T',title='Any'),
    alt.Y('sum(Quantitat):Q',title='Quantitat'),
    alt.Color('NomComarca:N',title='Comarca'),
    alt.Tooltip(['NomComarca:N','year(Any):T','sum(Quantitat):Q'])
).properties(
    height=350,
    width=750
).configure_axis(
    grid=False
).configure_view(
    strokeOpacity=0
).interactive()
```
